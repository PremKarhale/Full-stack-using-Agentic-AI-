```python
#==================================
# Step 3: Load PDF content using PyMuPDFLoader
#==================================

from langchain_community.document_loaders import PyMuPDFLoader

loader = PyMuPDFLoader(file_name)

docs = loader.load()
```

```python
print(docs)
print(len(docs))
```

```python
full_text= ""

for doc in docs:
  full_text += doc.page_content + "\n"

print(full_text)
```

```python
!pip install -q langchain_text_splitters
```

```python
#==================================
# Step 4: Split text into smaller chunks
#==================================


from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=900,
    chunk_overlap=180,
    separators=["\n\n", "\n", ".", " "]
)

text_chunks = splitter.split_text(full_text)

for chunk in text_chunks:
  print("=="*20)
  print(chunk)
print(len(text_chunks))
```

```python
#==================================
# Step 5: Convert text chunks into Document objects
#==================================


from langchain_core.documents import Document

chunks = [Document(page_content=chunk) for chunk in text_chunks]


# chunks =[]
# for chunk in text_chunks:

#   new_chunk = Document(page_content=chunk)
#   chunks.append(new_chunk)

# chunks
```

```python
print(chunks)
print(len(chunks))
```

```python
#==================================
# Step 6: Create FAISS vector store with GEMINI embeddings
#==================================
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_community.vectorstores import FAISS
from google.colab import userdata


gemini_key = userdata.get('GEMINI_API_KEY')

embedding_model = GoogleGenerativeAIEmbeddings(
    model="text-embedding-004",
    google_api_key=gemini_key
)


vectore_store = FAISS.from_documents(documents=chunks,embedding=embedding_model)
```

```python
#==================================
# Step 7: Perform similarity search (test serch)
#==================================

results = vectore_store.similarity_search("explain the three prong defence mechanism ") #test query

print(results)
print(len(results)) # gives me "revelent document " based on the query
```

```python
#==================================
# Step 8: Create a retriever for document retrieval
#==================================

retriever =vectore_store.as_retriever(   # just invoke the retriver by passing document and it gives relevent context
    search_type="similarity",
    search_kwargs={"k":4}
)
```

```python
# func to get the final context

def get_context_string(documents):
  context_text= ""

  for doc in documents:
    context_text += doc.page_content + "\n"

  return context_text
```

```python
print(get_context_string(results))
```

```python
#==================================
# Step 10: Initialize the Chat Model and Prompt Template
#==================================

from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate

llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite",api_key= gemini_key)

prompt = PromptTemplate(
    input_variables= ["question","context"],
    template="""
    you are a Helpful Ai Assitant.
    Answer ONLY using the following context from the uploaded PDF.
    If the context is insufficient, just say "I don't know".

    Context:
    {context}

    Question:
    {question}
    """
)
```

```python
#==================================
# Step 11: Create the LLM chain
#==================================
# Combine the prompt and model into a processing chain

chain = prompt | llm
```

```python
#==================================
# Step 12: Run an initial test query
#==================================

query = "Name the multilevel defence system to prevent the ZERO DAY DDOS ATTACKS"


retrieved_docs = retriever.invoke(query)
print(retrieved_docs) # returns relevent context

final_context = get_context_string(retrieved_docs)

print(final_context)
```

```python
response = chain.invoke({
    'question': query,
    'context': final_context
})
```

```python
print(response.content)
```

```python
# Build a eg chat model :
while True:


  user_input = input("You: ")

  if user_input.lower() in ["exit","bye"]:
    print("Goodbye!")
    break

  retrieved_docs = retriever.invoke(user_input)
  final_context = get_context_string(retrieved_docs)

  response = chain.invoke({
    'question': user_input,
    'context': final_context
  })


  print(f"Gemini: {response.text}")
```
