# Text Splitters
- Divide the Jraw data into chunks

```python
!pip install langchain_text_splitters -qU
```

```python
text = """
My name is Aisha.
I am 28 years old.
I live in Bangalore.
What do you do?
"""
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter


splitter = RecursiveCharacterTextSplitter(
    chunk_size=40,
    chunk_overlap=0
)

chunks = splitter.split_text(text)

for chunk in chunks:
  print("=="*20)
  print(chunk)
print(len(chunks))
```

# Embeddings
- convert raw data into Embeddings

```python
!pip install -qU google-genai
```

```python
text = """Artificial Intelligence (AI) is transforming the world.
It enables machines to learn, reason, and assist humans in numerous fields.
From healthcare to education, AI is everywhere.
LangChain helps developers build with LLMs easily and at scale."""
```

```python
from google.colab import userdata
api_key = userdata.get('gemini_key')
from google import genai


client = genai.Client(api_key = api_key)
result = client.models.embed_content(
     model="gemini-embedding-001",
     contents=text,
     config={
         "output_dimensionality": 16
     }
)

print(result.embeddings[0].values)
print(len(result.embeddings[0].values))
```

```python
from langchain_openai import OpenAIEmbeddings

embedding_model = OpenAIEmbeddings(api_key=openai_key,model="text-embedding-3-small")
```

```python
from langchain_core.documents import Document

doc1 = Document(page_content="Shah Rukh Khan, known as the King of Bollywood, is celebrated for his romantic roles and strong screen presence. He has starred in numerous iconic films like Dilwale Dulhania Le Jayenge and Chennai Express.")
doc2 = Document(page_content="Aamir Khan is known for his perfectionism and thought-provoking films. He has delivered memorable performances in movies like Lagaan, Dangal, and Taare Zameen Par, often focusing on social issues.")
doc3 = Document(page_content="Salman Khan is famous for his action-packed roles and mass appeal. Known for movies like Bajrangi Bhaijaan and Dabangg, he also hosts the popular reality show Bigg Boss.")
doc4 = Document(page_content="Ranbir Kapoor is admired for his versatility and charm. With films like Rockstar, Barfi, and Sanju, he has earned critical acclaim and a huge fan following among the youth.")
doc5 = Document(page_content="Deepika Padukone is one of the most successful actresses in Bollywood. Known for her powerful performances in films like Padmaavat and Piku, she has also worked in international projects.")

docs = [doc1, doc2, doc3, doc4, doc5]
```

```python
!pip install -q langchain_community faiss-cpu langchain
```

```python
from langchain_community.vectorstores import FAISS

vector_store = FAISS.from_documents(
    embedding=embedding_model,
    documents=docs
)
result = vector_store.similarity_search(query="Which actor is known for fighting movies",k=2)

print(result)
```

```python
# ---------------------------------------
# Step-1: Install the packages
# ---------------------------------------

!pip install langchain langchain_chroma chromadb langchain_openai langchain_google_genai -q
```

```python
# ---------------------------------------
# Step-2: Create documents or get documents
# ---------------------------------------

from langchain_core.documents import Document

doc1 = Document(page_content="Shah Rukh Khan, known as the King of Bollywood, is celebrated for his romantic roles and strong screen presence. He has starred in numerous iconic films like Dilwale Dulhania Le Jayenge and Chennai Express.")
doc2 = Document(page_content="Aamir Khan is known for his perfectionism and thought-provoking films. He has delivered memorable performances in movies like Lagaan, Dangal, and Taare Zameen Par, often focusing on social issues.")
doc3 = Document(page_content="Salman Khan is famous for his action-packed roles and mass appeal. Known for movies like Bajrangi Bhaijaan and Dabangg, he also hosts the popular reality show Bigg Boss.")
doc4 = Document(page_content="Ranbir Kapoor is admired for his versatility and charm. With films like Rockstar, Barfi, and Sanju, he has earned critical acclaim and a huge fan following among the youth.")
doc5 = Document(page_content="Deepika Padukone is one of the most successful actresses in Bollywood. Known for her powerful performances in films like Padmaavat and Piku, she has also worked in international projects.")

docs = [doc1, doc2, doc3, doc4, doc5]
```

```python
# ---------------------------------------
# Step-3: Create the Chroma Store
# ---------------------------------------
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_chroma import Chroma

from google.colab import userdata

gemini_key = userdata.get("gemini_key")
chroma_key = userdata.get("Chroma")
tenant_key = userdata.get("tenant")

embedding_model = GoogleGenerativeAIEmbeddings(api_key = gemini_key, model="text-embedding-004")


# Local Storage/In-memory Storage
vector_store = Chroma(
    collection_name="sample",
    persist_directory="chroma_db_2",
    embedding_function=embedding_model

)

# Cloud Stroage

vector_cloud_store = Chroma(
    collection_name="bollywood_actors_collection",
    embedding_function=embedding_model,
    database="pip_rag_base",
    chroma_cloud_api_key=chroma_key,
    tenant=tenant_key
)

# ---------------------------------------
# Step-4: Add documents
# ---------------------------------------


vector_data = vector_cloud_store.add_documents(docs)

print(vector_data)
```

```python
result=vector_cloud_store.similarity_search(query="Which actor is known for fighting movies",k=2)
print(result)
for r in result:
  print(r.page_content)
```
