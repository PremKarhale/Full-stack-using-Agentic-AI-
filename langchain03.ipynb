**Multiple chains**
- Parallel chains (Runnable Parallel)

```python
text = 'Unployement in India'
```

```python
def get_word_count(text):
    word_count=len(text.split())
    return word_count
def get_char_count(text):
    space_replace =text.replace(' ','')
    return len(space_replace)

print(get_word_count(text))
print(get_char_count(text))
```

- normal func cannot be run parallely , they must be done Runnablelambda(func()) to make them parallely invoked 

<!-- Runnablelamdba(func 1) | Runnablelambda(func 2) -->

```python
from langchain_core.runnables import RunnableLambda
# word_count = RunnableLambda(get_word_count)
# print(word_count.invoke(text))
```

```python
word_count = RunnableLambda(get_word_count)
char_count = RunnableLambda(get_char_count)
```

```python
from langchain_core.runnables import RunnableParallel
parallel = RunnableParallel({'words':word_count,'char':char_count})
result=parallel.invoke(text)
print(result)
```

- **Using Tavily for web search**


```python
import os 
from tavily import TavilyClient
print('api_key ',os.getenv('TAVELY_API_KEY'))
def web_search(query):
    client = TavilyClient(api_key=os.getenv('TAVELY_API_KEY'))
    response = client.search(
    query=query,
    include_answer="basic"
    )
    return (response.get('answer'))
result=web_search('who is the pm of india')
print(result)

# Whatever the ans i get in return i pass it to the RunnableParalles as a input 
```
