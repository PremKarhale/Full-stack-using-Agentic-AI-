```python
user_query='Messi visit to india '
```

```python
import os 
from tavily import TavilyClient
# print('api_key ',os.getenv('TAVILY_API_KEY'))

def web_search(query):
    client = TavilyClient(api_key=os.getenv('TAVELY_API_KEY'))
    response = client.search(
    query=query,
    include_answer="basic"
    )
    return (response.get('answer'))
web_result=web_search(user_query)
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(model='gemini-2.5-flash-lite',google_api_key=os.getenv('GEMINI_API_KEY'))
print(llm.invoke('my name is prem '))
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

fb_prompt = PromptTemplate(template="""
You are a professional Social Media Manager at a FAANG company with deep expertise in brand communication and audience engagement.

TASK:
Create a high-quality, engaging Facebook post based on the information provided below.

CONTEXT:
{web_results}

GUIDELINES:
- Write in a clear, polished, and conversational tone suitable for a broad Facebook audience.
- Ensure the post is engaging, positive, and value-driven.
- Highlight the most important and interesting insights from the content.
- Keep the length concise (3â€“5 sentences).
- Use line breaks for readability where appropriate.
- Include relevant emojis sparingly to enhance engagement.
- Add 2â€“4 relevant hashtags at the end.
- Avoid clickbait, exaggeration, or unverified claims.
- Do not mention that the content was generated by AI.

OUTPUT:
Return only the final Facebook post text. Do not include explanations or metadata.
""",input_variables=['web_results'])
fb_chain=fb_prompt | llm| StrOutputParser()

```

## twitter Chain

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

twitter_prompt = PromptTemplate(template="""
You are a professional Social Media Manager at a FAANG company with deep expertise in brand communication and audience engagement.

TASK:
Create a high-quality, engaging Facebook post based on the information provided below.

CONTEXT:
{web_results}

GUIDELINES:
- Write in a clear, polished, and conversational tone suitable for a broad Facebook audience.
- Ensure the post is engaging, positive, and value-driven.
- Highlight the most important and interesting insights from the content.
- Keep the length concise (3â€“5 sentences).
- Use line breaks for readability where appropriate.
- Include relevant emojis sparingly to enhance engagement.
- Add 2â€“4 relevant hashtags at the end.
- Avoid clickbait, exaggeration, or unverified claims.
- Do not mention that the content was generated by AI.

OUTPUT:
Return only the final Facebook post text. Do not include explanations or metadata.
""",input_variables=['web_results'])
twitter_chain=twitter_prompt | llm| StrOutputParser()

```

## RunnableParallel

```python
from langchain_core.runnables import RunnableParallel
resp = RunnableParallel(
{'Fb_post':fb_chain,
 'Linkdin_post':linkdin_chain,
 'twitter_post':twitter_chain
 }
)
print(resp.invoke({'web_results':web_result}))
```
