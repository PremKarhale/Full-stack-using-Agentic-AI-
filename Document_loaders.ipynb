#Text Loader

```python
!pip install langchain langchain_google_genai langchain_community -q
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
print(uploaded)
```

```python
file_name = list(uploaded.keys())[0]
```

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader(file_name)
```

```python
docs = loader.load()
```

```python
print(docs)
```

```python
context = docs[0].page_content
```

```python
print(context)
```

```python
from google.colab import userdata
gemini_key = userdata.get('GEMINI')
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(model='gemini-2.5-flash',api_key = gemini_key)

from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate(
    template="Write 10 ponts summary on the following context - {context}",
    input_variables=["context"]
)
```

```python
chain  = prompt | llm

response  = chain.invoke({'context': context})

print(response.content)
```

# PyMuPDF Loader

```python
!pip install langchain langchain_google_genai langchain_community -q
```

```python
!pip install -qU pymupdf
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
file_name = list(uploaded.keys())[0]
```

```python
print(file_name)
```

```python
from langchain_community.document_loaders import PyMuPDFLoader

loader = PyMuPDFLoader(file_name)
```

```python
docs = loader.lazy_load()
```

```python
print(len(docs))
```

```python
print(docs[1].page_content)
```

```python
full_text = ""
for doc in docs:
  full_text = doc.page_content + "\n\n"

  print(full_text)
```

```python
!pip install langchain langchain_google_genai langchain_community -q
```

```python
from google.colab import files

uploaded = files.upload()
```

```python
file_name = list(uploaded.keys())[0]
```

```python
print(docs)
print(len(docs))
```

```python
print(docs[0].page_content)
```

```python
print(docs[1].page_content)
```

```python
full_text = ""
for doc in docs:
  full_text = doc.page_content + "\n\n"

  print(full_text)
```
