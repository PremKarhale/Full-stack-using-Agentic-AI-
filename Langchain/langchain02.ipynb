Langchain using PromptTemplate

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
import os

while True :
  language = input('Enter any Programming language :')
  if language.lower() =='exit':
    print('ok Bye')
    break;

  llm = ChatGoogleGenerativeAI(model='gemini-2.5-flash-lite',google_api_key=os.getenv('GEMINI_API_KEY')) # initialize llm 
  joke_template=PromptTemplate(template='Give me 5 an joke on {lang}',input_variable=['lang'])

  #joke_template ->llm 
  #chains
  joke_chain = joke_template | llm |StrOutputParser()

  # invoke
  response = joke_chain.invoke({"lang":language})
  print(response)
```
