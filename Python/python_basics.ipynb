Question-4: return vs print â€” Build a Reusable Discount System Create two functions for a GenAI subscription discount.

Function-1: discount_print(price, discount_percent)

Calculates discounted price and prints it Does not return anything Function-2: discount_return(price, discount_percent)

Calculates discounted price and returns it Does not print anything Formula: final = price - (price * discount_percent / 100)

Proof code (must do):

a = discount_print(1000, 10)

b = discount_return(1000, 10)

Print a and b

Add comments explaining:

why a is None why b is 900

````python
def discount_print(price,discount_percent):
    final = price - (price * discount_percent / 100)
    print("final Discounted Price :",final)

def discount_return(price,discount_percent):
    final = price - (price * discount_percent / 100)
    return final

a=discount_print(1000,10)
b=discount_return(1000,10)
print(a) # value of a is none because we did not return anything 
print(b)

````

Question-5: Lambda Tasks for Prompt Processing
Given:

prompts = ["Write a poem", "Summarize", "Generate code", "Hi", "Explain AI", "Ok"]
Tasks:

Use map() + lambda to create a list of lengths of each prompt.
Use filter() + lambda to create a list of prompts with length > 5.
Print both lists.

````python
prompts=["Write a poem", "Summarize", "Generate code", "Hi", "Explain AI", "Ok"]
prompt_length=list(map(lambda x:len(x),prompts))
print(prompt_length)
prompt_filter = list(filter(lambda x: x>5,prompt_length ))
print(prompt_filter)
````
